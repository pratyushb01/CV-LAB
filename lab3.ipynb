4. Write a program to detect edges in a image.

```python

image = cv2.imread("stop.jpg")  
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Step 2: Apply Sobel operator in X and Y
sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=3)  # Horizontal edges
sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=3)  # Vertical edges

# Convert gradients to uint8
sobelx_abs = cv2.convertScaleAbs(sobelx)
sobely_abs = cv2.convertScaleAbs(sobely)

# Combine both directions
sobel_combined = cv2.addWeighted(sobelx_abs, 0.5, sobely_abs, 0.5, 0)

#Laplacian operator
laplacian = cv2.Laplacian(gray, cv2.CV_64F)
laplacian_abs = cv2.convertScaleAbs(laplacian)

plt.figure(figsize=(12, 6))

plt.subplot(2, 2, 1)
plt.title("Original Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(2, 2, 2)
plt.title("Sobel Edge Detection")
plt.imshow(sobel_combined, cmap='gray')
plt.axis("off")

plt.subplot(2, 2, 3)
plt.title("Sobel X")
plt.imshow(sobelx_abs, cmap='gray')
plt.axis("off")

plt.subplot(2, 2, 4)
plt.title("Laplacian Edge Detection")
plt.imshow(laplacian_abs, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```

4. Implement Canny edge detection algorithm.

```python


image = cv2.imread("stop.jpg")  
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

#  Apply Gaussian Blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 1.4)

#  Apply Canny edge detector
# Parameters: (image, lower_threshold, upper_threshold)
edges = cv2.Canny(blurred, 100, 200)

# Display original and edge-detected image
plt.figure(figsize=(10, 4))

plt.subplot(1, 2, 1)
plt.title("Original Grayscale")
plt.imshow(gray, cmap='gray')
plt.axis("off")

plt.subplot(1, 2, 2)
plt.title("Canny Edge Detection")
plt.imshow(edges, cmap='gray')
plt.axis("off")

plt.tight_layout()
plt.show()
```
