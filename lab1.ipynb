```python
videofile=cv2.VideoCapture(r'c:\Users\pratyush\Downloads\Video-220.mp4')

fps = videofile.get(cv2.CAP_PROP_FPS)
print(f"Frames per second: {fps}")


while videofile.isOpened():
    ret,frame=videofile.read()
    if not ret:
        break
    cv2.imshow('Video window',frame)

    if cv2.waitKey(33) &0xFF ==ord('q'):
        break
videofile.release()
cv2.destroyAllWindows()
```

```python
import cv2
from matplotlib import pyplot as plt
from IPython.display import clear_output
import time

cap = cv2.VideoCapture(r'c:\Users\pratyush\Downloads\Video-194.mp4')

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Convert from BGR to RGB
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Clear previous output
    clear_output(wait=True)
    plt.imshow(frame)
    plt.axis('off')
    plt.show()

    # Control frame rate
    time.sleep(0.03)  # ~30 FPS

cap.release()
```

```python

image=cv2.imread(r'c:\Users\pratyush\Downloads\daryl_crossbow.jpg')

def pixel_rgb_value(event,x,y,flags,param):
    if event == cv2.EVENT_LBUTTONDOWN:
        b,g,r = image[y,x]
        print(f"Clicked Pixed at ({x},{y}) . RGB values: R:{r},G:{g},B:{b}")

cv2.namedWindow('Daryl-action')
cv2.setMouseCallback('Daryl-action',pixel_rgb_value)

while True:
    cv2.imshow('Daryl-action',image)
    if cv2.waitKey(1) & 0xFF == 27:
        break
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

canvas=np.ones((500,500,3),dtype=np.uint8)*255

start=(100,100)
end=(300,400)
color=(255,0,0)
thickness=2

cv2.rectangle(canvas,start,end,color,thickness)

cv2.imshow('Drawn Rectanglee',canvas)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
image=cv2.imread(r'c:\Users\pratyush\Downloads\daryl2.jpg')

h,w=image.shape[:2]

center=(h//2,w//2)
angle=270
rotation_matrix=cv2.getRotationMatrix2D(center,angle,scale=1.0)
rotated_image=cv2.warpAffine(image,rotation_matrix,(w,h))

cv2.imshow("Original Image",image)
cv2.imshow("Rotated Image",rotated_image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
